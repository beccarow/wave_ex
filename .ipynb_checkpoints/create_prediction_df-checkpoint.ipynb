```python
categories = ['Low', 'Medium', 'High']
def binary(cat):
    binaryList = []
    
    for x in categories:
        if x in cat:
            binaryList.append(1)
        else:
            binaryList.append(0)
    
    return binaryList
```

```python
df['danceability_bin'] = df['danceability'].apply(lambda x: binary(x))
df['valence_bin'] = df['valence'].apply(lambda x: binary(x))
df['energy_bin'] = df['energy'].apply(lambda x: binary(x))
```

```python
artist_list = []
for index, row in spot.iterrows():
    artists = row["artist(s)_name"]
    
    for artist in artists:
        if artist not in artist_list:
            artist_list.append(artist)
len(artist_list)
```

```python
def binary_art(artist):
    binaryList = []
    
    for x in artist_list:
        if x in artist:
            binaryList.append(1)
        else:
            binaryList.append(0)
    
    return binaryList
```

```python
df['artist_bin'] = df['artist(s)_name'].apply(lambda x: binary_art(x))
```

```python
def getNeighbors(query_song, K):
        distances = []
    
        for index, song in song_df.iterrows():
            if song['id'] != query_song['id']:
                dist = Similarity(query_song['id'], song['id'])
                distances.append((song['id'], song['track_name'], dist))
    
        distances.sort(key=operator.itemgetter(2))
        neighbors = []
    
        for x in range(K):
            neighbors.append(distances[x])
        return neighbors
```

```python
for i in song_df.id.values:
    query_song = song_df.iloc[i]
    neighbors = getNeighbors(query_song, K=5)
    rec_titles = [neighbors[i][1] for i in range(len(neighbors))]
    song_df.loc[[i], 'recommendations'] = pd.Series([rec_titles], index=df.index[[i]])
```

```python
def Similarity(
    song1: int, 
    song2: int,
): # where song1 and song2 are the unique identifiers
    a = song_df.iloc[song1]
    b = song_df.iloc[song2]
    
    artistA = a['artist_bin']
    artistB = b['artist_bin']
    
    artistDistance = spatial.distance.cosine(artistA, artistB)
    
    danceA = a['danceability_bin']
    danceB = b['danceability_bin']
    danceDistance = spatial.distance.cosine(danceA, danceB)
    
    valenceA = a['valence_bin']
    valenceB = b['valence_bin']
    valenceDistance = spatial.distance.cosine(valenceA, valenceB)
    
    energyA = a['energy_bin']
    energyB = b['energy_bin']
    energyDistance = spatial.distance.cosine(energyA, energyB)
    
    return artistDistance + danceDistance + valenceDistance + energyDistance
```
